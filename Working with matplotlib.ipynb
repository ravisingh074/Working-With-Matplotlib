```python
# Import the pyplot module from matplotlib for creating visualizations
# Import numpy for numerical operations, particularly for working with arrays  
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the x & y coordinates of the points  
x_point = np.array([5,4,6,3,9])
y_point = np.array([9,8,6,5,0])
# Create a line plot with x_point and y_point
plt.plot(x_point,y_point)
plt.show()
```

```python
# Define the x & y coordinates of the points
x_point = np.array([5,4,6,3,9])
y_point = np.array([9,8,6,5,0])
# Create a scatter plot using 'o' to denote circular markers  
plt.plot(x_point,y_point,"o")
plt.show()
```

```python
x_point = np.array([5,4,6,3,9])
y_point = np.array([9,8,6,5,0])
# Create a scatter plot with circular markers, setting the marker size to 20, edge color to blue ('b'), and face color to green ('g')  
plt.plot(x_point, y_point, marker="o", ms=20, mec="b", mfc="g")  
plt.plot(x_point,y_point,marker = "o",ms= 20, mec = "b",mfc = "g")
plt.show()
```

```python
x_point = np.array([1,4,9,3,9])
y_point = np.array([0,8,0,5,0])
# Create a line plot with dashed lines ('--'), star markers ('*'), line width of 8, and green color ('g')  
plt.plot(x_point,y_point,linestyle="--",marker = "*", linewidth = "8", color ="g")
plt.show()
```

```python
x_point = np.array([1,4,1,3,9])
y_point = np.array([6,8,6,5,0])
plt.plot(x_point)
plt.plot(y_point)
plt.show()
```

```python
# Define the x-coordinates of the points  
x_point = np.array([0, 4, 6, 3, 5])  
# Define the y-coordinates of the points  
y_point = np.array([2, 5, 8, 2, 3])  
# Plot the x and y points  
plt.plot(x_point, label='X Point Data')   
plt.plot(y_point, label='Y Point Data')  
# Labeling the axes with specified color and size    
plt.ylabel("Oxygen", color="black", size="20")  
plt.xlabel("Air", color="black", size="20")    
# Title of the plot with specified font family, color, and size  
plt.title("Oxygen & Air label", family="serif", color="b", size="20")  
# Display the legend to identify data lines (optional, remove if not needed)  
plt.legend()   
plt.show()  
```

```python
x_point = np.array([1,4,1,3,9])
y_point = np.array([6,8,6,5,0])
xx_point = np.array([0,4,6,3,5])
yy_point = np.array([2,5,8,2,3])
plt.scatter(x_point,y_point, color ='g')
plt.scatter(xx_point,yy_point)
plt.show()
```

```python
x_point = np.array([1,4,1,3,9])
y_point = np.array([6,8,6,5,0])
colors = np.array([10,20,30,40,50])
plt.scatter(x_point, y_point,c= colors, cmap = 'viridis')
plt.colorbar()
plt.xlabel('X Point')  
plt.ylabel('Y Point')  
plt.title('Colored Scatter Plot')  
plt.show()
```

```python
x_point = np.array([1,4,1,3,9])
y_point = np.array([6,8,6,5,0])
plt.plot(x_point)
plt.plot(y_point)
plt.grid()
plt.show()
```

```python
x_point = np.array([1,4,1,3,9])
y_point = np.array([6,8,6,5,0])
plt.plot(x_point)
plt.plot(y_point)
plt.grid(axis = 'x')
plt.show()
```

```python
# creating histograms looks
x_point = np.array([1,4,1,3,9])
y_point = np.array([6,8,6,5,0])
plt.hist(x_point, bins = 5)
plt.hist(y_point, bins = 5)
plt.show()
```

```python
x_point = np.array([1,4,1,3])
# Create a pie chart with specified labels for each segment
plt.pie(x_point, labels = ["apple","banana","mango","cherry"])
# Set title for the pie chart  
plt.title('Fruit Distribution')  
plt.show()
```
